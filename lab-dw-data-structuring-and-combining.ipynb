```python
# Your code goes here

import pandas as pd
df_1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df_1
```

```python
df_1.rename(columns={col: col.replace(" ", "_").lower() for col in df_1.columns}, inplace=True)
df_1.rename(columns = {"st":"state"}, inplace = True)


df_1.columns
```

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df_2
```

```python
df_2.rename(columns={col: col.replace(" ", "_").lower() for col in df_2.columns}, inplace=True)
df_2.rename(columns = {"st":"state"}, inplace = True)

df_2.columns
```

```python
df_3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
df_3
```

```python
df_3.rename(columns={col: col.replace(" ", "_").lower() for col in df_3.columns}, inplace=True)
df_3.rename(columns = {"st":"state"}, inplace = True)

df_3.columns
```

```python
combined_df = pd.concat([df_1, df_2, df_3])
combined_df
```

```python
stand_gender = {"Male":"M", "m":"M", "M":"M", "F":"F", "female":"F", "Female":"F", "f":"F", "Femal":"F"}

combined_df["gender"] = combined_df["gender"].map(stand_gender).str.strip().str.upper()

combined_df["gender"].unique()
```

```python
state_replacements = {
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington"
    # Add other states as needed
}

combined_df['state'] = combined_df['state'].map(state_replacements)
combined_df['state'].unique()
```

```python
# Standardizing the 'education' column
combined_df['education'] = combined_df['education'].str.replace('Bachelors', 'Bachelor', regex=False)

# Cleaning the 'customer_lifetime_value' column
combined_df['customer_lifetime_value'] = combined_df['customer_lifetime_value'].str.replace('%', '', regex=False).astype(float)

# Standardizing the 'vehicle_class' column
vehicle_class_replacements = {
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
}
combined_df['vehicle_class'] = combined_df['vehicle_class'].replace(vehicle_class_replacements)

# Verify the changes
combined_df.head()
```

```python
combined_df["customer_lifetime_value"] = combined_df["customer_lifetime_value"].astype(float)

combined_df["customer_lifetime_value"]
combined_df.select_dtypes(include='number')
```

```python
combined_df["number_of_open_complaints"] = combined_df["number_of_open_complaints"].astype(str)
combined_df["number_of_open_complaints"] = combined_df["number_of_open_complaints"].apply(lambda x: x.split("/")[1] if "/" in x else x)
combined_df["number_of_open_complaints"].unique()
```

```python
import numpy as np
combined_df['number_of_open_complaints'] = combined_df['number_of_open_complaints'].apply(lambda x: int(x) if x.isdigit() else np.nan)
combined_df['number_of_open_complaints'].unique()
```

```python
combined_df.isna().sum()
```

```python
combined_df["gender"].isna().sum()
```

```python

num_nan = combined_df['gender'].isna().sum() # Calculate the number of NaN values in the 'Sex' column


if num_nan == 3054:
    num_male = num_nan // 2
    num_female = num_nan - num_male  # In case the number is odd, give the extra one to 'F'


    replacement_values = ['M'] * num_male + ['F'] * num_female #Create a list with 'M' and 'F' values
    np.random.shuffle(replacement_values)    # Shuffle the list to randomize the distribution
    combined_df.loc[combined_df['gender'].isna(), 'gender'] = replacement_values  #  Replace NaN values in the 'Sex' column with the shuffled 'M' and 'F' values

# Display the first few rows to verify

#print("Number of Male (M):", (insurance_df['gender'] == 'M').sum())
#print("Number of Female (F):", (insurance_df['gender'] == 'F').sum())

combined_df["gender"].isna().sum()
combined_df["gender"].value_counts()

#distributing Nan between F and M, because removing all rows is too much
```

```python
combined_df.dropna(subset = ["customer"], inplace = True)
combined_df.isna().sum()
```

```python
combined_df["customer_lifetime_value"] = combined_df["customer_lifetime_value"].fillna(combined_df["customer_lifetime_value"].mean())
combined_df["customer_lifetime_value"].isna().sum()
```

```python
combined_df["state"] = combined_df["state"].fillna (method = "ffill").fillna (method = "bfill")
combined_df["state"].value_counts()
```

```python
#Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer

pivot_table = df_4.pivot_table(
    values='customer_lifetime_value',
    index='sales_channel',
    aggfunc='sum'
)
pivot_table
```
